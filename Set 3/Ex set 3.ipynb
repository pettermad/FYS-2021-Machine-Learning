```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
mu1 = np.array([0, 0])
mu2 = np.array([1, 1])
sigma1 = np.eye(2)
sigma2 = np.eye(2)
w = np.array([1, -1])
w0 = 0

# Generate random samples from the two normal distributions
samples1 = np.random.multivariate_normal(mu1, sigma1, 5)
samples2 = np.random.multivariate_normal(mu2, sigma2, 5)

# Plot the samples
plt.scatter(samples1[:, 0], samples1[:, 1], c='red', label='Class 1')
plt.scatter(samples2[:, 0], samples2[:, 1], c='blue', label='Class 2')

# Plot the decision boundary
x = np.linspace(-2, 3, 100)
y = (-w[0] * x - w0) / w[1]
plt.plot(x, y, c='black', label='Decision Boundary')

# Set the axis labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

# Define the parameters
mu1 = np.array([0, 0])
mu2 = np.array([1, 1])
sigma1 = np.eye(2)
sigma2 = np.eye(2)
w = np.array([1, -1])
w0 = 0

# Generate random samples from the two normal distributions
samples1 = np.random.multivariate_normal(mu1, sigma1, 5)
samples2 = np.random.multivariate_normal(mu2, sigma2, 5)

# Calculate the sigmoid output for each sample
z = 1 / (1 + np.exp(-(np.dot(samples1, w) + w0)))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the samples
ax.scatter(samples1[:, 0], samples1[:, 1], z, c='red', label='Class 1')
ax.scatter(samples2[:, 0], samples2[:, 1], z, c='blue', label='Class 2')

# Plot the decision boundary
x = np.linspace(-2, 3, 100)
y = (-w[0] * x - w0) / w[1]
z_boundary = 0.5 * np.ones_like(x)
ax.plot(x, y, z_boundary, c='black', label='Decision Boundary')

# Set the axis labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

# Show the plot
plt.show()
```
